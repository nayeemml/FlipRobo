```python
 


import pandas as pd
from bs4 import BeautifulSoup
import requests


page=requests.get("https://www.imdb.com/chart/top/")


page.content



soup=BeautifulSoup(page.content,'html.parser')


print(soup.prettify())




movie_titles=soup.find_all('td', class_='titleColumn')

movie_ratings=soup.find_all('td',class_='ratingColumn imdbRating')


movie_name=[]
movie_year=[]
movie_rating=[]
for i in range(0,100):
    movie_name.append(movie_titles[i].a.text)
    movie_year.append(movie_titles[i].span.text)
    movie_rating.append(movie_ratings[i].strong.text)



imdb=pd.DataFrame({})

imdb["Name of the Movie"]=movie_name
imdb["Year of the Movie"]=movie_year
imdb["Movie Rating"]=movie_rating




imdb.index+=1


imdb



for i in range(1,100):
    imdb['Year of the Movie'][i]=imdb["Year of the Movie"][i][1:-1]


imdb["Year of the Movie"]



imdb



imdb.to_csv("imdb.csv")






```

```python
#!/usr/bin/env python
# coding: utf-8



# In[1]:


import pandas as pd
from bs4 import BeautifulSoup
import requests


# In[2]:


page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")


# In[3]:


page.content


# In[4]:


soup=BeautifulSoup(page.content,'html.parser')


# In[5]:


print(soup.prettify())


# In[6]:


movie_titles=soup.find_all('td', class_='titleColumn')

movie_ratings=soup.find_all('td',class_='ratingColumn imdbRating')


# In[7]:


movie_name=[]
movie_year=[]
movie_rating=[]
for i in range(0,100):
    movie_name.append(movie_titles[i].a.text)
    movie_year.append(movie_titles[i].span.text)
    movie_rating.append(movie_ratings[i].strong.text)


# In[8]:


imdb=pd.DataFrame({})

imdb["Name of the Movie"]=movie_name
imdb["Year of the Movie"]=movie_year
imdb["Movie Rating"]=movie_rating


# In[9]:


imdb


# In[10]:


imdb.index+=1


# In[11]:


imdb


# In[12]:


for i in range(1,100):
    imdb['Year of the Movie'][i]=imdb["Year of the Movie"][i][1:-1]


# In[13]:


imdb["Year of the Movie"]


# In[14]:


imdb


# In[15]:


imdb.to_csv("imdb-Top_100_Indian_Movies.csv")


# In[ ]:



```

```python
#!/usr/bin/env python
# coding: utf-8

# # Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
# # i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating. 

# In[1]:


import selenium
import pandas as pd
from selenium import webdriver


# In[2]:


driver = webdriver.Chrome(r"C:\Users\Home\Downloads\Nayeem\FlipRobo internship\WebSCraper\chromedriver.exe")


# In[3]:


url= "https://www.icc-cricket.com/rankings/womens/team-rankings/odi"


# In[4]:


driver.get(url)


# In[5]:


team_tags=driver.find_elements_by_xpath("//span[@class='u-hide-phablet']")
pos_tags=driver.find_elements_by_xpath("//td[@class='table-body__cell table-body__cell--position u-text-right']")
match_tags=driver.find_elements_by_xpath("//td[3][@class='table-body__cell u-center-text']")
points_tags=driver.find_elements_by_xpath("//td[4][@class='table-body__cell u-center-text']")
rating_tags=driver.find_elements_by_xpath("//td[@class='table-body__cell u-text-right rating']")


# In[6]:


teams=[]
position=[]
matches=[]
ratings=[]
points=[]

first_position=driver.find_element_by_xpath("//td[@class='rankings-block__banner--pos']")
first_matches=driver.find_element_by_xpath("//td[@class='rankings-block__banner--matches']")
first_points=driver.find_element_by_xpath("//td[@class='rankings-block__banner--points']")
first_ratings=driver.find_element_by_xpath("//td[@class='rankings-block__banner--rating u-text-right']")

position.append(first_position.text)
matches.append(first_matches.text)
ratings.append(first_ratings.text)
points.append(first_points.text)


# In[7]:



for i in team_tags:
    teams.append(i.text)

for i in pos_tags:
    position.append(i.text)

for i in match_tags:
    matches.append(i.text)
    
for i in rating_tags:
    ratings.append(i.text)

for i in points_tags:
    points.append(i.text)


# In[8]:


Top10Teams=pd.DataFrame({})


# In[9]:


Top10Teams["Position"]=position[0:10]
Top10Teams["Team Name"]=teams[0:10]
Top10Teams["Matches"]=matches[0:10]
Top10Teams["Points"]=points[0:10]
Top10Teams["Ratings"]=ratings[0:10]


# In[10]:


Top10Teams


# In[ ]:




```

# Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
#!/usr/bin/env python
# coding: utf-8

# # Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description. 

# In[1]:


import selenium
import pandas as pd
from selenium import webdriver


# In[2]:


driver = webdriver.Chrome(r"C:\Users\Home\Downloads\Nayeem\FlipRobo internship\WebSCraper\chromedriver.exe")


# In[3]:


url="https://forecast.weather.gov/MapClick.php?lat=37.7749&lon=-122.4194"


# In[4]:


driver.get(url)


# In[5]:


period_tag=driver.find_elements_by_xpath("//p[@class='period-name']")
short_desc=driver.find_elements_by_xpath("//p[@class='short-desc']")
temp_high=driver.find_elements_by_xpath("//p[@class='temp temp-high']")
temp_low=driver.find_elements_by_xpath("//p[@class='temp temp-low']")
long_desc=driver.find_elements_by_xpath("//div[@class='col-sm-10 forecast-text']")


# In[6]:


detailed_description=[]

for i in long_desc:
    detailed_description.append(i.text)
detailed_description


# In[ ]:





# In[7]:


temperature=[]

for i in temp_high:
    temperature.append(i.text)
for i in temp_low:
    temperature.append(i.text)
temperature


# In[8]:


day=[]
for i in period_tag:
    day.append(i.text.replace("\n"," "))
desc=[]
for i in short_desc:
    desc.append(i.text.replace("\n"," "))


# In[9]:


print(len(day),len(desc),len(temperature),len(detailed_description))


# In[10]:


Weather=pd.DataFrame({})

Weather["Period"]=day[0:8]
Weather["Short Description"]=desc[0:8]
Weather["Temperature"]=temperature[0:8]
Weather["Detailed Description"]=detailed_description[0:8]


# In[11]:


Weather


# In[ ]:




```

# Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.


```python
#!/usr/bin/env python
# coding: utf-8

# # Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.

# In[1]:


import selenium
import pandas as pd
from selenium import webdriver


# In[2]:


driver = webdriver.Chrome(r"C:\Users\Home\Downloads\Nayeem\FlipRobo internship\WebSCraper\chromedriver.exe")


# In[3]:


url = "https://www.monsterindia.com/srp/results?query=software%20developer%20&searchId=a43d3d34-8201-4f81-b729-147bce86d184"


# In[4]:


driver.get(url)
import time


# In[5]:


job=[]
company=[]
location=[]
experience=[]


# In[6]:


location=[]
for i in range(0,5):
    job_tags=driver.find_elements_by_xpath("//*[@id='srp-right-part']/div/div[1]/div[1]/div[2]/div/div/div/div[1]/div/div/h3/a")

    for i in job_tags:
        job.append(i.text)
    #########
    company_tags=driver.find_elements_by_xpath("//span[@class='company-name']")

    for i in company_tags:
        company.append(i.text)
    ########
    location_tags=driver.find_elements_by_xpath("//*[@id='srp-right-part']/div/div[1]/div[1]/div[2]/div/div/div/div/div/div/div/div[1]/span/small")
    
    for i in location_tags:
        location.append(i.text)
    
    #########
    experience_tags=driver.find_elements_by_xpath("//*[@id='srp-right-part']/div/div[1]/div[1]/div[2]/div/div/div/div/div/div/div/div[2]/span/small")
    
    for i in experience_tags:
        experience.append(i.text)

    driver.find_element_by_xpath("//button[@class='btn-next-prev btn-next']").click()
    time.sleep(5)


# In[7]:


len(job),len(company),len(location),len(experience)


# In[8]:


jobinCompany=pd.DataFrame({})

jobinCompany["Job Title"]=job
jobinCompany["Company name"]=company
jobinCompany["location"]=location
jobinCompany["experience"]=experience


# In[9]:


jobinCompany


# In[ ]:




```
